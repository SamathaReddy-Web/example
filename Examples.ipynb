```python
from transformers import pipeline
generator=pipeline("text-generation", model="gpt2")
prompt="Explain difference between LangChain and Langraph"   # zero shot
```

```python
output = generator(prompt, max_length=100, num_return_sequences=1)
print(output[0]['generated_text'])
```

```python
from transformers import pipeline
translation=pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-tc-big-en-es")
prompt=""" Translate English to Spanish :
English: Good morning
Spanish: Buen dia
English: Thank You
Spanish:                """
print(translation(prompt))
```

```python
from transformers import pipeline
calc=pipeline("text2text-generation", model="google/flan-t5-large")
prompt="Ram initially have 8 apples. He gave away 3 apples to his friend. How many apples does Ram have now?"
print(calc(prompt))
```
